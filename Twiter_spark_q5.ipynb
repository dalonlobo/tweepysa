```python
from __future__ import print_function
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
from pprint import pprint
from pyspark.sql import SparkSession
from pyspark import SparkContext

```

```python
sc = SparkContext(appName="StreamingTwitterAnalysis")
sc.setLogLevel("Error")
ssc = StreamingContext(sc,30)
```

```python
ssc.checkpoint("checkpoint")
socket_stream = ssc.socketTextStream("172.31.60.179",9094)
```

```python
lines = socket_stream.window(360)
#lines.pprint()
```

```python
def parse(line):
    try:
        return [(line, 1)]
    except:
        return []
```

```python
users = lines.flatMap(parse).reduceByKey(lambda a,b :a+b)
users_count= users.transform((lambda foo:foo.sortBy(lambda x:( -x[1]))))
```
